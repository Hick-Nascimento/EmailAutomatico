```python
import win32com.client as client
import pandas as pd
import datetime as dt
```

```python
tabela = pd.read_excel('contas a Receber.xlsx')
display(tabela)
tabela.info()
```

```python
hoje = dt.datetime.now()
print(hoje)
```

```python
tabela_devedores = tabela.loc[tabela['Status']=='Em aberto']
display(tabela_devedores)
tabela_devedores = tabela_devedores.loc[tabela_devedores['Data Prevista para pagamento']>hoje]
display(tabela_devedores)
```

```python
outlook = client.Dispacth('Outlook.Application')
```

```python
emissor = outlook.session.Accounts['xxxxx@hotmail.com']
```

```python
mensagem = outlook.CreateItem(0)
mensagem.Display()
messagem.To = "XXXX@gmail.com"
mensagem.Subject = 'Atraso no Pagamento'
mensagem.Body = '''

ALGUMA COISA AQUI

'''

mensagem._ole_.Invoke(*(64209,0,8,0,emissor))
mensagem.Save()
mensagem.Send()
```

```python
dados = tabela_devedores[['Valor em aberto', 'Data Prevista para pagamento','E-mail','NF']].values.tolist()
```

```python
for dado in dados:
    destinatario = dado[2]
    nf = dado[3]
    prazo = dado[1]
    prazo = prazo.strftime("%d/%m/%y")
    valor = dado[0]
    
    mensagem = outlook.CreateItem(0)
    mensagem.Display()
    messagem.To = destinatario
    mensagem.Subject = 'Atraso no Pagamento'
    mensagem.Body = f'''

   Prezado Cliente,

   verificamos um atraso no pagamento referente a NF {nf} com vencimento em {prazo} e valor total de R${valor:.2f}.
   Gostariamos de verificar se hÃ¡ algum problema que necessite de auxilio de nossa equipe.

   Em caso de dÃºvidas, Ã© sÃ³ entrar em contato com nosso time atrÃ¡ves do e-mail tecobreiautomaticamentecompython.

   Att,
   
'''

    mensagem._ole_.Invoke(*(64209,0,8,0,emissor))
    mensagem.Save()
    mensagem.Send()
```
